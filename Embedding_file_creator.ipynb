```python

import os
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
import gradio as gr

def process_pdfs(pdf_files):
    raw_text = ""
    for pdf_file in pdf_files:
        pdf_reader = PdfReader(pdf_file)
        for page in pdf_reader.pages:
            raw_text += page.extract_text()

    text_chunks = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20).split_text(raw_text)
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2", model_kwargs={'device': 'cuda'})
    vector_store = FAISS.from_texts(text_chunks, embeddings)
    vector_store.save_local("Embeddings")
    return "PDFs processed and embeddings saved!"

iface = gr.Interface(
    fn=process_pdfs,
    inputs=gr.Files(),  # Accept multiple files
    outputs="text",
    title="PDF Processing and Embedding",
    description="Upload PDF files to extract text, create chunks, and generate embeddings.",
)

iface.launch()
```
